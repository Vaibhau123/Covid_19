# Forecast for confirmed cases

```python
confirmd.columns = ['ds', 'y']
confirmd
```

```python
# import libraries

from prophet import Prophet

```

```python
m = Prophet()         # assigning value Prophet  to variable 'm'
```

```python
m.fit(confirmd)     # fit data inside
```

```python
future = m.make_future_dataframe(periods = 7)
```

```python
future
```

```python
forecast = m.predict(future)
```
